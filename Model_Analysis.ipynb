```python
# Suppression and Elevation Analysis - Normal Conductance

mli_alone = pickle.load(open('run1_GPF_GMLI_0', 'rb'))
pf_2_mli_1 = pickle.load(open('run1_GPF_GMLI_2', 'rb'))
pf_4_mli_1 = pickle.load(open('run1_GPF_GMLI_4', 'rb'))
pf_6_mli_1 = pickle.load(open('run1_GPF_GMLI_6', 'rb'))

all_runs = [mli_alone, pf_2_mli_1, pf_4_mli_1, pf_6_mli_1]


for i in range(len(all_runs)):
    all_runs[i] = all_runs[i] - 200

suppression = np.array([])
mean_mod_ratio = np.array([])
min_mod_ratio = np.array([])
baseline_mean = np.array([])
max_mod_ratio = np.array([])

for i in range(len(all_runs)):
    counts_raw, bins = np.histogram(all_runs[i], np.linspace(-200,100,151))
    counts = np.array(counts_raw/(num_trials +1))

    idx1 = np.where(bins == -50); idx1 = idx1[0].astype('int')
    idx2 = np.where(bins == 20); idx2 = idx2[0].astype('int')
    len_array = idx2-idx1
    data = np.zeros([len_array[0],2])
    data[:,0] = bins[idx1[0]:idx2[0]]
    data[:,1] = counts[idx1[0]:idx2[0]]


    baseline_mean = np.append(baseline_mean, np.mean(data[0:25,1]))
    suppression = np.append(suppression, np.min(data[26:31,1]))
    mean_mod_ratio = np.append(mean_mod_ratio,np.mean(data[26:31,1]) / np.mean(data[0:25,1]))
    min_mod_ratio = np.append(min_mod_ratio,np.min(data[26:31,1]) / np.mean(data[0:25,1]))
    max_mod_ratio = np.append(max_mod_ratio,np.max(data[26:31,1] / np.mean(data[0:25,1])))

mod_depth_minimum = (1 - min_mod_ratio)* -1
mod_depth_maximum = (1 - max_mod_ratio)* -1

summary_data =  {'Model':['0 pf:0 mli1', '2 pf:1 mli', '4 pf:1 mli', '6 pf:1 mli'],
           'Baseline Firing Rate':baseline_mean * 500, 
           'Minimum Firing Rate': suppression *500,
           'Modulation Depth - Min (% Change)': mod_depth_minimum * 100, 
           'Modulation Depth - Max (% Change)': mod_depth_maximum * 100, 
          }

data_frame = pd.DataFrame.from_dict(summary_data)
data_frame = data_frame.set_index('Model')  

norm_conductance = np.array([mod_depth_minimum, mod_depth_maximum])

data_frame

```
